```python
# filtering out indoor AQ sensors...
la_cleaning5 = la_cleaning4[la_cleaning4['location_type']=='outside']

# checking if any sensors have been offline for a while...
la_cleaning5['last_seen'].sort_values().head(10)
```

```python
# let's drop the first 2 since the sensors haven't been online in the last month
la_cleaning6 = la_cleaning5.drop([516, 579])       #['last_seen'].sort_values().head(10)
```

```python
la_cleaning6.drop(columns=['location_type'], inplace=True)
```

```python
la_cleaning6.info()
```

```python
# YAY, no nulls!

la_cleaning6.isna().sum()
```

```python
la_clean = la_cleaning6.groupby('fips_id').mean()
la_clean.head(3)
```

```python
# Correlation of AQ with other features

AQ_corrs = pd.DataFrame(la_clean.corr()['1week_avg']).sort_values(by='1week_avg')
AQ_corrs
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = la_clean[['TotPop', 'Ac_Water', 'Ac_Land', 'Ac_Unpr', 'Pct_AO0', 'Pct_AO1', 'Pct_AO2p', 'NatWalkInd']]
y = la_clean['1week_avg']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

ss=StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)

dummy = DummyRegressor(strategy="mean")
dummy.fit(X_train_scaled, y_train)
preds_dum = dummy.predict(X_test_scaled)
print('Dummy training score: ', dummy.score(X_train_scaled, y_train).round(4))
print('Dummy testing score: ', dummy.score(X_test_scaled, y_test).round(4))
print('MSE: ', mean_squared_error(y_test, preds_dum, squared=False).round(4))
```

```python
lr = LinearRegression(n_jobs=-1)
lr.fit(X_train_scaled, y_train)
preds_lr = lr.predict(X_test_scaled)

print('LR training score: ', lr.score(X_train_scaled, y_train).round(4))
print('LR testing score: ', lr.score(X_test_scaled, y_test).round(4))
print('MSE: ', mean_squared_error(y_test, preds_lr, squared=False).round(4))
```
